```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
#Read 1st csv files as pandas dataframe and view
csv_file = "Raw_Data/Property_Sale.csv"
PS_df = pd.read_csv(csv_file)
PS_df.head()
```

```python
#Split 1st dataframe to drop columns that are not needed
#1st dataframe
#First part of the split
PS_1st_df =PS_df.iloc[:, 0:2]
PS_1st_df.head()
```

```python
#1st dataframe
#Second part of the split
#Droped columns that are not needed
PS_2nd_df = PS_df.loc[:, 'state':'property_type']
PS_2nd_df.head()
```

```python
#Concatenate 1st dataframe
PS_new = [PS_1st_df, PS_2nd_df]
PS_final_df = pd.concat(PS_new, axis=1)
#Rename columns of 1st dataframe
PS_final_df = PS_final_df.rename(columns={"date_sold": "Date_Sold", \
    "price": "Property_Price", "suburb": "Suburb", "state": "State", \
    "lat": "Latitude", "lon": "Longitude", "bedrooms": "Bedrooms", "property_type": "Property_Type"})

#display
PS_final_df.head()
```

```python
# Dropping the rows with "NaN"
# identify partial string
PS_final_df = PS_final_df.dropna()
  
#display
PS_final_df
```
